# Historical Oil Supply Disruptions Comparison

Comparing the magnitude of major geopolitical oil supply disruptions since 1973.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'crisis': [
        'Yom Kippur\nWar (1973)',
        'Iranian\nRevolution (1979)',
        'Iran-Iraq\nWar (1980)',
        'Gulf War\n(1990)',
        'Strait of Hormuz\nClosure (2026)'
    ],
    'supply_disruption_pct': [6, 4, 4, 6, 20]
}

df = pd.DataFrame(data)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

colors = ['#2c3e50'] * 4 + ['#e74c3c']

bars = ax.bar(df['crisis'], df['supply_disruption_pct'], 
              color=colors, width=0.5, edgecolor='white')

for bar, val in zip(bars, df['supply_disruption_pct']):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
            f'{val}%', ha='center', fontsize=11, fontweight='bold')

ax.set_title('Historical Oil Supply Disruptions: How 2026 Compares',
             fontsize=14, fontweight='bold', pad=20)
ax.set_xlabel('Crisis', fontsize=11)
ax.set_ylabel('Global Oil Supply Removed (%)', fontsize=11)
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('../visualizations/03_historical_oil_crises.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Sources
- 1973, 1979, 1980, 1990 disruption figures: Federal Reserve Bank of Dallas, 
  "What the closure of the Strait of Hormuz means for the global economy" (March 2026)
- 2026 figure: IEA, International Energy Agency strategic reserve release statement (March 2026)

**Note:** 1973 and 1990 figures represent "a little more than 6%", shown as 6% for comparison.
All figures sourced directly from Dallas Fed (March 2026), cross-referenced with IEA.
